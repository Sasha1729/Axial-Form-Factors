```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(z):
  return (1/z)*(np.log(1 + z))
```

```python
x = np.linspace(-1, 5, 100)
y = func(x)

def pd(x, y, N, M):
    # Fitting polynomials to the data and finding the coefficients of the polynomials
    num_coeffs = np.polyfit(x, y, N)
    denom_coeffs = np.polyfit(x, y, M)
    return np.polyval(num_coeffs, x)/np.polyval(denom_coeffs, x)
```

```python
# Plotting the original function and its Pade approximation

p11 = (1 + (1/6)*x)/(1 + (2/3)*x)
p12 = (1 + (1/2)*x)/(1 + x + (1/6)*(x**2))
p22 = (1 + (7/10)*x + (1/30)*(x**2))/(1 + (6/5)*x + (3/10)*(x**2))

plt.plot(x, y, label = r'$f(x)$')
# plt.plot(x, pd(x, y, 1, 1), label = r'$P_{1}^{1}$')
# plt.plot(x, pd(x, y, 1, 2), label = r'$P_{1}^{2}$')
# plt.plot(x, pd(x, y, 2, 2), label = r'$P_{2}^{2}$')
plt.plot(x, p11, label = r'$P_{1}^{1}$')
plt.plot(x, p12, label = r'$P_{1}^{2}$')
plt.plot(x, p22, label = r'$P_{2}^{2}$')

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Pade Approximation')
plt.show()
```

#### $P_{0}^{1}$

```python
# Expansion (a0) + (-a0*b1)*q2 + (a0*b1^2)*q2^2 + Order(q2^3)
# [0.95664768 1.50612178]
r_a2 = 6*(0.95664768 * 1.50612178)
m01 = np.sqrt(12/r_a2)
r01 = np.sqrt(r_a2)
print('Radius:', r01)
print('Mass of nucleon:', m01)
```

#### $P_{1}^{1}$

```python
# Expansion (a0) + (-a0*b1 + a1)*q2 + (a0*b1^2 - a1*b1)*q2^2 + Order(q2^3)
# [0.96933602 0.09117386 1.73288426]

r_a2 = 6*(0.96933602 * 1.73288426 - 0.09117386)
m11 = np.sqrt(12/r_a2)
r11 = np.sqrt(r_a2)
print('Radius:', r11)
print('Mass of nucleon:', m11)
```

#### $P_{1}^{2}$
